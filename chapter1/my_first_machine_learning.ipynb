# 1. 나의 첫 머신러닝

## 1-1. 인공지능과 머신러닝, 딥러닝
* 인공지능이란
  - 인공지능 : 사람처럼 학습하고 추론할 수 있는 지능을 가진 컴퓨터 시스템을 만드는 기술
  - 역사
    * 1. 인공지능 태동기
      - 1943 : 워런 매컬러와 월터 피츠는 최초로 뇌의 뉴런 개념을 발표
      - 1950 : 앨런 튜링은 인공지능이 사람과 같은 지능을 가졌는지 테스트할 수 있는 유명한 튜링 테스트 발표
      - 1956 : 많은 과학자가 참여한 다트머스 AI 컨퍼런스에서는 인공지능에 대한 장밋빛 전망이 최고조
    * 2. 인공지능 황금기
      - 1957 : 프랑크 로젠블라트가 로지스틱 회귀의 초기 버전으로 볼 수 있는 퍼셉트론을 발표
      - 1959 : 데이비드 허블과 토르스텐 비셀이 고양이를 사용해 시각 피질에 있는 뉴런 기능응ㄹ 연구
    * 이후 컴퓨터 성능에 한계로 인해 AI 겨울이 도래
    * AI 겨울인 시기를 극복한 후 인공지능은 다시 각광을 받음
      - 1998 : 얀 르쿤이 신경망 모델(LeNet-5, 최초의 합성곱 신경망)을 만들어 손글씨 숫자를 인식하는 데 성공
      - 2012 : 제프리 힌턴의 팀이 이미지 분류 대회인 ImageNet에서 AlexNet(합성곱 신경망)으로 압도적인 우승

* 머신러닝이란
  - 머신러닝
    * 자동으로 데이터에서 규칙을 학습하는 알고리즘을 연구하는 분야
    * 인공지능의 하위 분야 중에서 지능을 구현하기 위한 소프트웨어
  - 사이킷런(scikit learn)
    * 컴퓨터 과학 분야의 대표적인 머신러닝 라이브러리
    * 해당 라이브러리는 파이썬 API를 사용
    * 성능이 검증된 머신러닝 알고리즘이 포함됨

* 딥러닝이란
  - 인공 신경망을 기반으로 한 방법들을 통칭
  - 오픈소스 머신러닝 라이브러리
    * 텐서플로
    * 파이토치

## 1-3. 마켓과 머신러닝
* 용어 정리
  - 특성
    * 데이터의 특징
    * 데이터를 표현하는 하나의 성질
  - 모델
    * 알고리즘이 구현된 객체
  - 정확도
    * 정확한 답을 몇 개 맞추었는지를 백분율로 나타낸 값
  - 분류
    * 여러 개의 클래스 중 하나를 구별해 내는 문제
  - 이진 분류
    * 2개의 클래스 중 하나를 고르는 문제

* 그래프 그리기
  - 맷플롯립(matplotlib)
    * 파이썬에서 과학계산용 그래프를 그리는 대표적인 패키지
    * scatter() : 산점도를 그리는 함수
      - c : 색 지정 (r, g, b, c, m, y, k, w 등)

* K-최근접 이웃 알고리즘 (KNN 알고리즘)
  - 주변에 가장 가까운 K개의 샘플을 통해 값을 예측하는 방식
  - 지도학습에 포함됨
  - sklearn.neighbors에 포함됨
    * KNeighborsClassifier() : 분류
    * KNeighborsRegressor() : 회귀
    * 매개변수
      - n_neighbors : 이웃의 개수 지정 (default 5)
      - p : 거리를 재는 방법 지정 (default 2)
        * 1 : 맨해튼 거리
        * 2 : 유클리디안 거리 
      - n_jobs : 사용할 CPU 코어 지정 (default 1)
        * -1 : 모든 CPU 코어 사용
        * 1 : CPU 코어 1 사용

* 머신러닝 관련 메서드
  - fit()
    * 사이킷런 모델을 훈련할 때 사용하는 메서드
    * 처음 두 매개변수로 훈련에 사용할 특성 데이터와 정답 데이터를 전달
  - predict()
    * 사이킷런 모델을 훈련하고 예측할 때 사용
    * 특성 데이터만 매개 변수로 전달
  - score()
    * 훈련된 사이킷런 모델의 성능 측정
    * 처음 두 매개변수로 특성과 정답 데이터를 전달
