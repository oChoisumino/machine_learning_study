# 2. 데이터 다루기

## 2-1. 훈련 세트와 테스트 세트
* 지도 학습과 비지도 학습
  - 머신러닝 알고리즘은 지도학습과 비지도 학습, 강화 학습으로 분류
  - 지도 학습
    * 입력과 타깃을 전달하여 모델을 훈련한 다음, 새로운 데이터를 예측하는 데 활용
    * 입력 데이터와 타깃 데이터가 훈련 데이터로 필요하다.
  - 비지도 학습
    * 타깃 없이 입력 데이터만 사용하여 어떤 특징을 찾는 데 주로 활용
    * 정답을 활용하지 않아 정답을 맞출 수는 없다.
    * 테이터를 잘 파악하거나 변형하는 데 도음을 준다.
  - 강화 학습
    * 타깃이 아니라 알고리즘이 행동한 결과로 얻은 보상을 사용하여 학습한다.

* 훈련 세트와 테스트 세트
  - 훈련 세트
    * 모델을 훈련할 때 사용할 데이터
    * 클수록 좋아 테스트 세트를 제외한 모든 데이터를 사용
  - 테스트 세트
    * 평가에 사용할 데이터
    * 전체 데이터에서 주로 20~30%를 테스트 세트로 사용
    * 데이터 양이 매우 많을 경우, 1%의 데이터만 테스트 데이터로 사용해도 무방할 수 있다.

* 샘플링 편향
  - 훈련세트에 특정 종류의 샘플이 치우쳐져 있는 현상
  - 훈련세트와 테스트 세트에 샘플이 골고루 섞이지 않은 경우 발생
  - 샘플링 편향을 가진 경우 제대로 된 지도학습 모델을 만들 수 없다.

* 넘파이
  - 파이썬의 대표적인 배열 라이브러리
  - 1차원 : 선, 2차원 : 면, 3차원 공간
  - seed()
    * 넘파이에서 난수를 생성하기 위한 정수 초깃값 지정
    * 초깃값이 동일하면 항상 결과가 동일
  - arrange()
    * 일정한 간격의 정수 또는 실수 배열을 만든다.

## 2-2. 데이터 전처리
* 넘파이로 데이터 준비하기
  - column_stack()
    * 전달받은 리스트를 일렬로 세운 다음 차례대로 나란히 연결
    * zip과 비슷한 역할을 하는 함수
    * 연결할 리스트는 파이썬 튜플로 전달
    * 예) [1, 1, 1], [0, 0, 0]인 배열 결과 [[1,0], [1,0], [1,0]]
  - ones(), zeros() 
    * 각각 원하는 개수의 1과 0을 채운 배열을 만드는 함수
  - concatenate()
    * 첫번째 차원을 따라 배열을 연결
    * 연결할 리스트는 파이썬 튜플로 전달
    * 예) [1, 1, 1], [0, 0, 0]인 배열 결과 [1, 1, 1, 0, 0, 0]

* 훈련 세트와 테스트 세트로 나누기
  - train_test_split()
    * 훈련 데이터를 훈련 세트와 테스트 세트로 나누는 함수
    * 매개 변수
      - random_state : 랜덤 시드 지정
      - stratify : 클래스 레이블이 담긴 배열을 전달하면 클래스 비율에 맞게 훈련 세트와 테스트 세트를 나눈다.
      - shuffle : 훈련 세트와 테스트 세트로 나누기 전 무작위로 섞을 지 여부 결정 (default True)
      - test_size : 테스트 세트로 나눌 비율 지정 (default 0.25)

* KNeighborsClassifier 클래스
  - 주변 샘플들의 클래스를 보고, 그 중의 다수 클래스를 골라서 반환
  - kneighbors() 메서드
    * 이웃까지의 거리와 이웃 샘플의 인덱스를 반환
    * n_neighbors 가 default 5 이므로 5개의  이웃이 반환됨

* 동일 기준 적용
  - x와 y 축 스케일이 다른 경우 거리를 이용한 알고리즘에서는 잘못된 결과를 이끌어낸다.
  - xlim, ylim
    * matplotlib 에서 x축과 y축 범위를 지정하여 표시

* 표준 점수
  - 가장 널리 사용하는 전처리 방법 중 하나
  - 훈련 세트의 스케일을 바꾸는 대표적인 방법 중 하나
  - 각 특성값이 평균에서 표준편차의 몇 배만큼 떨어져 있는 지를 나타냄
  - $표준 점수 = \frac{(특성값 - 평균)}{표준편차}$

* 분산
  - 데이터에서 평균을 뺀 값을 제곱한 다음 평균을 낸 값

* 표준 편차
  - 분산의 제곱근
  - 데이터가 분산된 정도를 나타냄

* 넘파이에서 제공하는 통계값 관련 함수
  - mean() : 평균을 계산 (axis 0 : 열, 1: 행 의 통계값을 계산)
  - std() : 표준편차를 계산 (axis 0 : 열, 1: 행 의 통계값을 계산)

* 브로드 캐스팅
  - 크기가 다른 넘파이 배열에서 자동으로 사칙연산을 모든 행이나 열로 확장하여 수행
  - 넘파이 배열 사이에서 일어남
